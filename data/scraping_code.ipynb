```python
import requests
from bs4 import BeautifulSoup
import csv
import time

kode_ptn = [
    1111,1112,1113,1114,1115,
    1121,1122,
    1131,1133,
    1141,1142,1143,
    1151,1161,1171,1181,
    1191,1192,
    1311,1321,1323,1324,
    1331,1332,1333,1334,1335,
    1341,1342,
    1351,1352,1353,1354,1355,1356,
    1361,1362,1363,1364,
    1371,1372,1373,
    1381,1382,1383,1384,1385,
    1511,1521,1531,
    1541,1542,
    1551,
    1611,1612,1613,
    1621,
    1631,1632,
    1711,1712,1718,1721,1722,1731,
    1741,
    1751,1752,1753,
    1811,1821,
    1911,1912,1913,1921
]

base_url = "https://sidatagrun-public-1076756628210.asia-southeast2.run.app/ptn_sb.php?ptn={}"
all_data = []
header = None
max_cols = 0

for kode in kode_ptn:
    ptn = str(kode)[1:]  # buang angka pertama
    url = base_url.format(ptn)
    print(f"Scraping {url} ...")

    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    table = soup.find("table")
    if table is None:
        print(f"❌ Tidak ada data untuk kode {kode} (ptn={ptn})")
        continue

    rows = table.find_all("tr")

    # ambil header dari tabel pertama
    if header is None:
        header = [h.text.strip() for h in rows[0].find_all(["th", "td"])]
        max_cols = len(header)

    for row in rows[1:]:
        cols = [col.text.strip() for col in row.find_all("td")]
        if cols:
            if len(cols) > max_cols:
                max_cols = len(cols)
            all_data.append([kode, ptn] + cols)

    time.sleep(1)

# pastikan semua baris sama panjang
for i in range(len(all_data)):
    diff = max_cols - (len(all_data[i]) - 2)  # kurangi 2 karena ada [kode, ptn]
    if diff > 0:
        all_data[i] += [""] * diff  # tambahin kolom kosong biar balance

# buat header final
if header:
    header_final = ["Kode 4 Digit", "PTN (3 digit)"] + header
    # kalau header kurang panjang → tambahin ColX
    if len(header) < max_cols:
        header_final += [f"Col{i}" for i in range(len(header)+1, max_cols+1)]
else:
    header_final = ["Kode 4 Digit", "PTN (3 digit)"] + [f"Col{i}" for i in range(1, max_cols+1)]

# simpan ke CSV
with open("hasil_sidatag.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(header_final)
    writer.writerows(all_data)

print("✅ Data berhasil disimpan ke hasil_sidatag.csv dengan", max_cols, "kolom")
```
